```python
from tkinter import filedialog
from tkinter import *
root = Tk()
root.ourCSVfile = filedialog.askopenfilename(initialdir = '/',
                                           title = 'Select File',
                                           filetypes = (("all file types", '*.*'),
                                                       ("coma seperated files", '*.csv'),
                                                       ("text files", '*.txt'))             
                                          )
root.destroy()
```

lets use pandas to make it a pandas dataframe

```python
from pandas import *
df = pandas.read_csv(root.ourCSVfile)
```

```python
Now lets plot with pandas
```

```python
df.plot(x ='x', y='x^2', kind = 'line')
```

```python
df.plot(x ='x', y='x^3/100', kind = 'line')
```

```python
df.plot(x ='x', y='abs(40*(x-10))', kind = 'line')
```

```python
import matplotlib.pyplot as plt

# gca stands for 'get current axis'
ax = plt.gca()

df.plot(kind='line',x='x',y='x^2',ax=ax)
df.plot(kind='line',x='x',y='x^3/100', color='red', ax=ax)
df.plot(kind='line',x='x',y='abs(40*(x-10))', color='yellow', ax=ax)

plt.show()
```
